```python
def scale_image(file_path, old_dir, new_dir, width, height):
    img = cv2.imread(file_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (width, height), interpolation = cv2.INTER_AREA)
    new_path = file_path.replace(old_dir, new_dir)
    cv2.imwrite(new_path, img)
    
def load_images(file_paths, num=0):
    imgs = None
    
    if num == 0: 
        num = len(file_paths)
        files_sel = file_paths
    else:
        for i in range(num):
            idx = np.random.choice(len(file_paths), 1)[0]
            files_sel.append(files[idx])
            
    for fname in files_sel:
        img = cv2.imread(fname)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
        if imgs is None:
            imgs = np.empty([0, img.shape[0], img.shape[1], img.shape[2]], dtype = np.uint8)
            
        imgs = np.append(imgs, img[None,:], axis=0)

    return imgs

def display_image_set(imgs, filename, cols_num=6, width=300):
    height_init, width_init, _ = imgs[0].shape
    
    if width_init > width:
        width_img = int(math.ceil(float(width) / float(cols_num)))
    else:
        width_img = width_init
        
    imgs_num = len(imgs)
    rows_num = int(math.ceil(float(imgs_num) / float(cols_num)))
    Kwidth = float(width_init) / float(width_img)
    height_img = int(height_init / Kwidth)
    result_shape = (rows_num * height_img, cols_num * width_img, 3)
    results_image = 255 * np.ones(shape=result_shape, dtype=np.uint8)

    for idx, img in enumerate(imgs):
        if img.shape[2] == 1:
            img = np.array(cv2.merge((img,img,img)),np.uint8)
            
        row = idx // cols_num
        col = idx - row * cols_num
        
        if width_init != width_img:
            img_resized = cv2.resize(img, (width_img, height_img), interpolation = cv2.INTER_AREA)
        else:
            img_resized = img
        
        h, w, _ = img_resized.shape
        origin_y = row * h
        origin_x = col * w
        results_image[origin_y:origin_y+h, origin_x:origin_x+w, :] = img_resized
        
    cv2.imwrite('./preview/{}.jpg'.format(filename), results_image)

    plt.figure(figsize=(25, 25))
    plt.imshow(results_image)
    plt.axis('off')
    plt.show()

#     plt.rcParams["figure.figsize"] = (25, 25)
#     plt.imshow(results_image.astype(np.uint8))
#     plt.axis('off') 
#     plt.show()
    
def rescale_dataset(files_dir, scaled_files_dir, width):
    train_file_paths = glob.glob(files_dir + '/**/*.jpeg')
    print("Train file paths: {}".format(len(train_file_paths)))

    for file_path in train_file_paths:
        scale_image(file_path, files_dir, scaled_files_dir, width, width)

def load_dataset(files_dir, num=0):
    file_paths = glob.glob(files_dir + '/**/*.jpeg')
    images = load_images(file_paths)
    print("Loaded dataset: {}".format(images.shape))
    
    return images
```

```python
batch_size = 12
epochs = 100
img_height = target_image_width
img_width = target_image_width
```

```python
train_image_generator = ImageDataGenerator(rescale=1./255)
train_data_gen = train_image_generator.flow_from_directory(
    batch_size=batch_size,
    directory=train_scaled_dir,
    shuffle=True,
    target_size=(img_height, img_width),
    class_mode='categorical'
)
```

```python
test_image_generator = ImageDataGenerator(rescale=1./255)  
test_data_gen = test_image_generator.flow_from_directory(
    batch_size=batch_size,
    directory=test_scaled_dir,
    shuffle=True,
    target_size=(img_height, img_width),
    class_mode='categorical'
)
```

```python
base_model = efn.EfficientNetB0(
    weights='imagenet', 
    include_top=False, 
    input_shape=(target_image_width, target_image_height, target_image_channels)
)
base_model.trainable = True
base_model.summary()
```

```python
classifier = keras.models.Sequential()
classifier.add(base_model)
classifier.add(Flatten())
classifier.add(Dense(2, activation='softmax'))
classifier.summary()
classifier.compile(
    optimizer=Adam(), 
    loss='categorical_crossentropy', 
    metrics=['accuracy']
)
```

```python
history = classifier.fit(
    train_data_gen, 
    epochs=epochs, 
    validation_data=test_data_gen
)
```

```python
classifier.save('covid-imagenet.h5')
```

```python
y = np.concatenate([test_data_gen.next()[1] for i in range(test_data_gen.__len__())])
true_labels = np.argmax(y, axis=-1)
prediction = classifier.predict(test_data_gen, verbose=2)
prediction = np.argmax(prediction, axis=-1)
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=85)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
            horizontalalignment="center",
            color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
def to_colorspace(img, space='LUV'):
    if space == 'HSV':
        return cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
    elif space == 'LUV':
        return cv2.cvtColor(img, cv2.COLOR_RGB2LUV)
    elif space == 'HLS':
        return cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    elif space == 'YUV':
        return cv2.cvtColor(img, cv2.COLOR_RGB2YUV)
    elif space == 'YCrCb':
        return cv2.cvtColor(img, cv2.COLOR_RGB2YCrCb)
    elif space == 'gray':
        return np.expand_dims(cv2.cvtColor(img, cv2.COLOR_RGB2GRAY), 2)
    else: 
        return img
    
def hog_previews(files):
    previews = []
    
    for file in files:   
        img = load_images([file])[0]
        previews.append(img)
        
        preview_channels = hog_features(img, vis=True, vec=True)
        previews.append(preview_channels)

    return previews

def normalize_gray_to_rgb(img):
    if (img.max() != 0):
        img *= 255.0 / img.max()
    else:
        img = 0
        
    img = np.expand_dims(img, 2)
    img = np.array(cv2.merge((img, img, img)), np.uint8)
    
    return img

def hog_features(img, space='LUV', orient=8, ppc=8, cpb=2, vec=True):
    features = []
    previews = []
    img = to_colorspace(img, space)
    red = img[:,:,0]
    
    feature_descriptors, preview = hog(
        red, 
        orientations=orient, 
        pixels_per_cell=(ppc, ppc), 
        cells_per_block=(cpb, cpb), 
        transform_sqrt=True, 
        visualize=True, 
        feature_vector=vec
    )

    features.append(feature_descriptors)
            
    preview = normalize_gray_to_rgb(preview)
    preview = exposure.rescale_intensity(preview, in_range=(0, 10)) 
    previews.append(preview)
    
    return [np.array(np.ravel(features)).astype(np.float64), np.array(previews).astype(np.uint8)]

def hog_previews(files_dir, num=6):
    file_paths = glob.glob(files_dir + '/**/*.jpeg')[:num]
    imgs = load_images(file_paths)
    
    previews = []
    fds = []
    
    for img in imgs:
        previews.append(img)
        fd, preview = hog_features(img)
        previews.extend(preview)
        fds.append(fd)
        
    previews_array = np.array(previews).astype(np.uint8)
    fds_array = np.array(fds).astype(np.float64)
    print("Sample descriptors {} , preview {}".format(fds_array.shape, previews_array.shape))
    display_image_set(previews, "HOG images")
```
