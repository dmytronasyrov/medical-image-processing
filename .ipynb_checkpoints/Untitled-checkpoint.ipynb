```python
def scale_image(file_path, old_dir, new_dir, width, height):
    img = cv2.imread(file_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (width, height), interpolation = cv2.INTER_AREA)
    new_path = file_path.replace(old_dir, new_dir)
    cv2.imwrite(new_path, img)
    
def load_images(file_paths, num=0):
    imgs = None
    files_sel = []
    
    if num == 0:
        num = len(file_paths)
        files_sel = file_paths
    else:
        for i in range(num):
            idx = np.random.choice(len(file_paths), 1)[0]
            files_sel.append(file_paths[idx])
            
    for fname in files_sel:
        img = cv2.imread(fname)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
        if imgs is None:
            imgs = np.empty([0, img.shape[0], img.shape[1], img.shape[2]], dtype = np.uint8)
            
        imgs = np.append(imgs, img[None,:], axis=0)

    return imgs

def display_image_set(imgs, filename, cols_num=6, width=1200):
    height_init, width_init, _ = imgs[0].shape
    
    if width_init > width:
        width_img = int(math.ceil(float(width) / float(cols_num)))
    else:
        width_img = width_init
        
    imgs_num = len(imgs)
    rows_num = int(math.ceil(float(imgs_num) / float(cols_num)))
    Kwidth = float(width_init) / float(width_img)
    height_img = int(height_init / Kwidth)
    result_shape = (rows_num * height_img, cols_num * width_img, 3)
    results_image = 255 * np.ones(shape=result_shape, dtype=np.uint8)

    for idx, img in enumerate(imgs):
        if img.shape[2] == 1:
            img = np.array(cv2.merge((img,img,img)),np.uint8)
            
        row = idx // cols_num
        col = idx - row * cols_num
        
        if width_init != width_img:
            img_resized = cv2.resize(img, (width_img, height_img), interpolation = cv2.INTER_AREA)
        else:
            img_resized = img
        
        h, w, _ = img_resized.shape
        origin_y = row * h
        origin_x = col * w
        results_image[origin_y:origin_y+h, origin_x:origin_x+w, :] = img_resized
        
    cv2.imwrite('./preview/{}.jpg'.format(filename), results_image)

    plt.figure(figsize=(25, 25))
    plt.imshow(results_image)
    plt.axis('off')
    plt.show()
    
def rescale_dataset(files_dir, scaled_files_dir, width):
    file_paths = glob.glob(files_dir + '/**/*.jpeg')
    print("File paths: {}".format(len(file_paths)))

    for file_path in file_paths:
        scale_image(file_path, files_dir, scaled_files_dir, width, width)

def load_dataset(files_dir, num=0):
    file_paths = glob.glob(files_dir + '/**/*.jpeg')
    images = load_images(file_paths, num=num)
    print("Loaded dataset: {}".format(images.shape))
    
    return images
```

```python
def image_data_generator(files_dir, batch_size, img_width, img_height, class_mode='categorical'):
    image_generator = ImageDataGenerator(rescale=1./255)
    
    return image_generator.flow_from_directory(
        batch_size=batch_size,
        directory=files_dir,
        shuffle=True,
        target_size=(img_height, img_width),
        class_mode=class_mode)

def draw_accuracy_graph(history):
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()
    
def draw_loss_graph(history):
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()

def draw_result(history):
    draw_accuracy_graph(history)
    draw_loss_graph(history)
    
def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion matrix', cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=85)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    thresh = cm.max() / 2.
    
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
            horizontalalignment="center",
            color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    
def metrics(val_data_gen, classifier):
    y = np.concatenate([val_data_gen.next()[1] for i in range(val_data_gen.__len__())])
    true_labels = np.argmax(y, axis=-1)

    prediction = classifier.predict(val_data_gen, verbose=2)
    prediction = np.argmax(prediction, axis=-1)

    cm = confusion_matrix(y_true=true_labels, y_pred=prediction)

    cm_plot_labels = ['COVID','Normal','Viral Pneumonia']
    plot_confusion_matrix(cm=cm, classes=cm_plot_labels, title='Confusion Matrix')
    acc = accuracy_score(true_labels, prediction) 
    print('Accuracy: %.2f%%' % (acc*100))

    precision = precision_score(true_labels, prediction,labels=[1,2], average='micro')
    print('Precision: %.2f%%' % (precision*100))

    recall = recall_score(true_labels, prediction, average='micro')
    print('Recall: %.2f%%' % (recall*100))

    score = f1_score(true_labels, prediction, average='micro')
    print('F-Measure: %.2f%%' % (score*100))

    tn, fp, fn, tp = cm.ravel()
    print("Sensitivity %.2f%%" % ((tp / (tp + fn))*100))
    print("Specificity %.2f%%" % ((tn / (tn + fp))*100))
    
    report = classification_report(true_labels, prediction, target_names=['normal','pneumonia'])
    print("Report", report)
```

```python
def to_colorspace(img, space='LUV'):
    if space == 'HSV':
        return cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
    elif space == 'LUV':
        return cv2.cvtColor(img, cv2.COLOR_RGB2LUV)
    elif space == 'HLS':
        return cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    elif space == 'YUV':
        return cv2.cvtColor(img, cv2.COLOR_RGB2YUV)
    elif space == 'YCrCb':
        return cv2.cvtColor(img, cv2.COLOR_RGB2YCrCb)
    elif space == 'gray':
        return np.expand_dims(cv2.cvtColor(img, cv2.COLOR_RGB2GRAY), 2)
    else: 
        return img
    
def hog_previews(files):
    previews = []
    
    for file in files:   
        img = load_images([file])[0]
        previews.append(img)
        
        preview_channels = hog_features(img)
        previews.append(preview_channels)

    return previews

def normalize_gray_to_rgb(img):
    if (img.max() != 0):
        img *= 255.0 / img.max()
    else:
        img = 0
        
    img = np.expand_dims(img, 2)
    img = np.array(cv2.merge((img, img, img)), np.uint8)
    
    return img

def hog_features(img, space='LUV'):
    img = to_colorspace(img, space)
    red = img[:,:,0]
    
    feature_descriptors, preview = hog(
        red, 
        orientations=8, 
        pixels_per_cell=(10, 10), 
        cells_per_block=(4, 4), 
        block_norm='L2',
        transform_sqrt=False, 
        visualize=True, 
        feature_vector=False
    )
            
    preview = normalize_gray_to_rgb(preview)
    preview = exposure.rescale_intensity(preview, in_range=(0, 10)) 
    
    return [np.ravel(feature_descriptors), preview]

def extract_hog(files_dir, hog_dir=None, num=0):
    file_paths = glob.glob(files_dir + '/**/*.jpeg')
    files_sel = []
    imgs = []
    fds = []
    previews = []
    files = []
    
    if num == 0: 
        num = len(file_paths)
        files_sel = file_paths
    else:
        for i in range(num):
            idx = np.random.choice(len(file_paths), 1)[0]
            files_sel.append(file_paths[idx])

    for fname in files_sel:
        files.append(fname)
        
        img = load_images([fname])[0]
        imgs.append(img)
        
        fd, preview = hog_features(img)
        fds.append(fd)
        previews.append(preview)
        
        if (hog_dir != None):
            hog_img = np.array(preview).astype(np.uint8)
            new_path = fname.replace(files_dir, hog_dir)
            cv2.imwrite(new_path, preview)
            cv2.imshow("image", preview)
    
    file_paths_np = np.array(files)
    imgs_np = np.array(imgs).astype(np.uint8)
    previews_np = np.array(previews).astype(np.uint8)
    fds_np = np.array(fds).astype(np.float64)

    return [fds_np, previews_np, imgs_np, file_paths_np]
    
def hog_previews(files_dir, num=6):
    fds, hogs, imgs, file_paths = extract_hog(files_dir, num=num)
    res = np.column_stack((imgs, hogs))
    print("Sample descriptors {} , HOGs {}, images {}, files {}".format(fds.shape, hogs.shape, imgs.shape, file_paths.shape))
    display_image_set(res, "HOG images")
    
def save_hog(files_dir, data_file, fds, hogs):
    file_paths_normal = glob.glob(files_dir + '/NORMAL/*.jpeg')
    file_paths_pneumonia = glob.glob(files_dir + '/PNEUMONIA/*.jpeg')

    labels_normal = np.ones(len(file_paths_normal))
    labels_pneumonia = np.zeros(len(file_paths_pneumonia))
    labels = np.concatenate((labels_normal, labels_pneumonia), axis = 0)
    print("Saving", fds.shape, hogs.shape, labels.shape, " to ", data_file + ".npz")
    np.savez_compressed(
        data_file,
        fds=fds, 
        hogs=hogs, 
        labels=labels
    )
```
